```python
sub_df = df['工资'].str.split('-|k|K',expand=True)
#sub_df = sub_df.str.extract('(\d*)')
#mean_salary = sub_df[:,1]
sub_df['new1'] = pd.to_numeric(sub_df[0])
sub_df['new2'] = pd.to_numeric(sub_df[2])
sub_df['mean'] = sub_df.mean(axis=1)
df['salary_mean'] = sub_df['mean']
df.head()
```

### 职位描述处理

```python
jd = df['职位描述']
st = []
for i in range(len(jd)):
    s = str(jd[i])
    n = s.find('要求')
    m = s.find('资格')
    if n > 0:
        t = s[n:].find('\n\n')
        st.append(s[n:t+n])
    elif m > 0:
        t = s[m:].find('\n\n')
        st.append(s[m:t+m])
    else:
        st.append([''])
df['职位要求'] = st      
```

### 词性标注

```python
from jieba import posseg
jd = df['职位要求']

jd_key = []
for i in jd:
    if len(i) > 1:
        for a, b in posseg.lcut(i):
            if b == "n" or b == "eng":
                jd_key.append(a)
```

```python
from collections import Counter
c = Counter(jd_key)
c.most_common()
```
